# Python + Pandas + Plotly on Finance Data to get Candlestick chart

### Read and Visualize S&P500 Stock Data

```python
# Get stock data for S&P500
from datetime import date
from pandas_datareader.data import DataReader
df = DataReader("^GSPC",'yahoo',start= date(2017,1,1))
#visualize the stock price trend
import plotly.graph_objs as go
trace = go.Candlestick(x=df.index,
                       open=df.Open,
                       high=df.High,
                       low=df.Low,
                       close=df.Close,
                       increasing=dict(name='up',line=dict(color= 'purple')),
                       decreasing=dict(name='down',line=dict(color= 'gray')))
layout = {
    'title': 'S&P500 Stock for 2017-present',
    'yaxis': {'title': 'Price in USD'}}
fig = dict(data=[trace], layout=layout)
import plotly.offline as ol
ol.init_notebook_mode(connected=True)
ol.iplot(fig)
```
