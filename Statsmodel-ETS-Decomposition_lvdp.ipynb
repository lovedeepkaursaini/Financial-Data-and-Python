## Statsmodel and seasonal_decompose to ETS

seasonal_decompose method has several parameters, one of them is model: additive or multiplicative or what?
- use an additive model when it seems that the trend is more linear and the seasonality and trend components seem to be constant over time (e.g. every year we add 10,000 passengers).
- use a multiplicative model when we are increasing (or decreasing) at a non-linear rate (e.g. each year we double the amount of passengers).

above chart shows the trend as slightly increasing at a higher rate than just linear, so we should try multiplicative model. but let's compare both these models below: this will show similar results for this example as we can notice from trend chart above which is quite at to edge of additive and multiplicative models.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result_mult = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')
result_add = seasonal_decompose(airline['Thousands of Passengers'], model='additive')
```

```python
#extract trend
result_mult.trend.plot()
```

```python
result_add.trend.plot()
```

```python
#get all the ETS plots
fig = result_mult.plot()
```
